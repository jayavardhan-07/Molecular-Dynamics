```python
import os
import matplotlib.pyplot as plt
from glob import glob
import pandas as pd
```

```python
data = sorted(glob("data_*.txt"))
append_data = pd.DataFrame(columns=["T","lat"])
```

```python
for x in range(len(data)):
    temp = data[x]
    df = pd.read_csv(temp, delimiter = ' ')
    df_m = df.mean()
    append_data.loc[x] = [df_m["T"], df_m["lat"]]
append_data.to_csv('lat_T.csv', index=False)
```

```python
plt.plot(append_data['T'], append_data['lat'],marker='o', linestyle = '-')
plt.savefig('lat_T.png')
```
